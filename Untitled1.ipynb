```python
# What are the names and marks of Five class students who has more than 50 mark?

file[(file["mark"] > 50) & (file["class"] == "Five")][["name","mark"]]
```

```python
'''

Average mark across all students

Average mark across male and female students

Max and min mark

Max mark for male and female students

Average mark for class 'Five'

Average mark for class 'Six' and male students only 

'''

print(file["mark"].mean())

print(file[file["gender"] == "male"]["mark"].mean())
print(file[file["gender"] == "female"]["mark"].mean())

print(file["mark"].max())
print(file["mark"].min())

print(file[file["gender"] == "male"]["mark"].min())
print(file[file["gender"] == "male"]["mark"].max())

print(file[file["gender"] == "female"]["mark"].min())
print(file[file["gender"] == "female"]["mark"].max())

print(file[file["class"] == "Five"]["mark"].mean())

print(file[(file["class"] == "Six") & (file["gender"] == "male")]["mark"].mean())
```
